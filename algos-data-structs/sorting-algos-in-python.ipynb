# Selection Sort


Pseudo Code for Selection Sort 

```
for i from 1 to n:
  minindex = i
  for j from i+1 to n:
    if A[j] < A[minindex]
      minindex = j
    {} 
    swap(A[i], A[minindex])

```

 

````python
def selection_sort(lst):
  
  for i in range(len(lst)):
    min_idx = i

    for j in range(i+1,len(lst)):
      if lst[j] < lst[min_idx] :
        min_idx = j
    
    lst[i],lst[min_idx] = lst[min_idx], lst[i]
  
  return lst
````

````python
lst = [5,4,3,6,2,1,7]

selection_sort(lst)
````

# Bubble Sort 

[Youtube link to understand](https://www.youtube.com/watch?v=nmhjrI-aW5o)

````python
def sort_binary_list(lst):
    """
    A function to sort binary list
    :param lst: A list containing binary numbers
    :return: A sorted binary list
    """
    
    size = len(lst)
    # Traverse through all array elements
    for i in range(size):
      for j in range(size-i-1):
        # traverse the array from 0 to n-i-1
        # Swap if the element found is greater
        # than the next element
        if lst[j] > lst[j+1]:
          lst[j], lst[j+1] = lst[j+1], lst[j]

    return lst
````

````python
lst = [1, 0, 1, 0, 1, 1, 0, 0]
````

````python
sort_binary_list(lst)
````

# Merge Sort

[Video Explanation](https://www.youtube.com/watch?v=JSceec-wEyw)

````python
def merge_sort(my_list):
    if len(my_list) > 1:
        mid = len(my_list) // 2
        left = my_list[:mid]
        right = my_list[mid:]

        # Recursive call on each half
        merge_sort(left)
        merge_sort(right)

        # Two iterators for traversing the two halves
        i = 0
        j = 0
        
        # Iterator for the main list
        k = 0
        
        while i < len(left) and j < len(right):
            if left[i] < right[j]:
              # The value from the left half has been used
              my_list[k] = left[i]
              # Move the iterator forward
              i += 1
            else:
                my_list[k] = right[j]
                j += 1
            # Move to the next slot
            k += 1

        # For all the remaining values
        while i < len(left):
            my_list[k] = left[i]
            i += 1
            k += 1

        while j < len(right):
            my_list[k]=right[j]
            j += 1
            k += 1
````
