# Recurssion

Recursion is the process in which a function calls itself during its execution. Each recursive call takes the program one scope deeper into the function.

The recursive calls stop at the base case. The base case is a check used to indicate that there should be no further recursion.

Imagine recursive calls as nested boxes where each box represents a function call. Each call makes a new box. When the base case is reached, we start moving out of the boxes one by one:


```python
def rec_count(num):
  # Base case
  if num == 0:
    return 0
  rec_count(num-1)
  print(num)


rec_count(5)
```

```python
def rec_count(number):
    print(number)
    # Base case
    if number == 0:
        return 0
    rec_count(number - 1)  # A recursive call with a different argument
    print(number)


rec_count(5)
```

```python
def grid_paths(n,m):
  if n==1 or m==1:
    return 1
  else:
    return grid_paths(n, m-1) + grid_paths(n-1,m)
```

```python
grid_paths(2,4)
```

# 5 Steps for Recussion
1. What's the simplestt possible input - Base Case
2. Play around with examples and visualize!
3. Relate hard cases to simpler cases
4. Genralize the pattern
5. Write code by combining recursive pattern with the base case 



```python
def count_partitions(n,m):
  if n == 0:
    return 1
  elif m == 0 or n < 0:
    return 0
  else:
    return count_partitions(n-m,m) + count_partitions(n, m-1)
```

```python
count_partitions(9,5)
```

# Factorial with Recurrsion

```python
def factorial(num):
  if num==0 or num==1:
    return 1
  elif num < 0:
    return -1
  else:
    return num*factorial(num-1) #recursive call 
```

```python
def factorial(num):
  if num ==0:
    return 1
  elif num < 0:
    return -1
  else:
    val = 1
    while num > 0:
      val *= num
      num -= 1
    return val
```

```python
factorial(5)
```
