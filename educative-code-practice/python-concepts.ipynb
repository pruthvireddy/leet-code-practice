## map() operator

The built-in map() function creates a map object using an existing list and a function as its parameters. This object can be converted to a list using the list() function (more on this later).The template for map() is as follows:

map(function, list)

The function will be applied, or mapped, to all the elements of the list.

Below, we’ll use map() to double the values of an existing list

```python
num_list = [0, 1, 2, 3, 4, 5]

double_list = map(lambda n: n * 2, num_list)

print(list(double_list))
```

## Filter()

Another similar example is the filter() function. It requires a function and a list.

filter() filters elements from a list if the elements satisfy the condition that is specified in the argument function.

Let’s write a filter() function that filters all the elements which are greater than 10

```python
numList = [30, 2, -15, 17, 9, 100]

greater_than_10 = list(filter(lambda n: n > 10, numList))
print(greater_than_10)
```

# For Loop w/ break

```python
n = 50
num_list = [10, 4, 23, 6, 18, 27, 47]
found = False  # This bool will become true once a pair is found

for n1 in num_list:
    for n2 in num_list:
        if(n1 + n2 == n):
            found = True  # Set found to True
            break  # Break inner loop if a pair is found
    if found:
        print(n1, n2) # Print the pair
        break  # Break outer loop if a pair is found
```

## Fibonacci

```python
n =7
fib_list = [0,1,1,2,3,5,8,13,21]
```

```python

type(fib_list)
```

```python
fib_list = list()
num=0
while num < n:
  if num ==0:
    fib_list.append(0)
  elif num ==1 or num ==2:
    fib_list.append(1)
  else:
    fib_list.append(fib_list[len(fib_list)-1]+ fib_list[len(fib_list)-2])

  num+=1

print(fib_list[len(fib_list)-1])
```
