# Balance checker for open and closed brackets

```python
def check_balance(brackets):  # The argument is a string
    
    op_br =0
    cl_br =0
    
    for i in range(0, len(brackets)):
      
      brac = brackets[i]

      if brac == '[':
        op_br +=1
        continue
      elif brac == ']':
        cl_br +=1
        continue
      
    if op_br == cl_br:
      return True
    else:
      return False
```

```python
def check_balance(brackets):
    check = 0
    for bracket in brackets:
        if bracket == '[':
            check += 1

        elif bracket == ']':
            check -= 1

        if check < 0:
            break

    return check == 0
```

```python
check_balance('[[[[]]]')
```

# Exercise: A Sum of Zero

You must implement the check_sum() function which takes in a list and returns True if the sum of two numbers in the list is zero. If no such pair exists, return False.

```python
def check_sum(num_list):
    for num1 in num_list:
      for num2 in num_list:
        if num1+num2 == 0:
          return True
          break
        else:
          continue
    return False
```

```python
def check_sum(num_list):
    for first_num in range(len(num_list)):
        for second_num in range(first_num + 1, len(num_list)):
            if num_list[first_num] + num_list[second_num] == 0:
                return True
    return False
```
