# Balance checker for open and closed brackets

```python
def check_balance(brackets):  # The argument is a string
    
    op_br =0
    cl_br =0
    
    for i in range(0, len(brackets)):
      
      brac = brackets[i]

      if brac == '[':
        op_br +=1
        continue
      elif brac == ']':
        cl_br +=1
        continue
      
    if op_br == cl_br:
      return True
    else:
      return False
```

```python
def check_balance(brackets):
    check = 0
    for bracket in brackets:
        if bracket == '[':
            check += 1

        elif bracket == ']':
            check -= 1

        if check < 0:
            break

    return check == 0
```

```python
check_balance('[[[[]]]')
```
